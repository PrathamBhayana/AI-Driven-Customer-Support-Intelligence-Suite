# Exploratory Data Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install xlrd
```

```python
file_path = '../data/ai_dev_assignment_tickets_complex_1000.xls'
tickets_df = pd.read_excel(file_path, engine="xlrd")
tickets_df.head()
```

```python
product_names = tickets_df['issue_type'].unique()
print(product_names)
```

```python
product_names = tickets_df['product'].unique()
print(product_names)
```

```python
missing_values = tickets_df.isnull().sum()
missing_values[missing_values > 0]
```

# Distribution of issue types

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=tickets_df, x='issue_type', order=tickets_df['issue_type'].value_counts().index)
plt.title('Distribution of Issue Types')
plt.xticks(rotation=45)
plt.show()
```

# The distribution of urgency levels

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=tickets_df, x='urgency_level', order=tickets_df['urgency_level'].value_counts().index)
plt.title('Distribution of Urgency Levels')
plt.xticks(rotation=45)
plt.show()
```

# The relationship between issue type and urgency level

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=tickets_df, x='issue_type', hue='urgency_level', order=tickets_df['issue_type'].value_counts().index)
plt.title('Issue Type vs Urgency Level')
plt.xticks(rotation=45)
plt.legend(title='Urgency Level')
plt.show()
```

# Products that generate the most support tickets

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=tickets_df, x='product', order=tickets_df['product'].value_counts().index)
plt.title('Top Products by Ticket Volume')
plt.xlabel('Product')
plt.ylabel('Number of Tickets')
plt.xticks(rotation=45)
plt.show()
```
